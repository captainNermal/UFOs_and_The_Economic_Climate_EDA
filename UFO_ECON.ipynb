## Introduction

In 1961, Frank Drake proposed the following equation to estimate the number of intelligent extraterrestrial societies (SETI, 2021):

$$N = R_{*} \cdot f_{p} \cdot n_{e} \cdot f_{L} \cdot f_{i} \cdot f_{c} \cdot L$$

$N$: Number of Extraterrestrial Civilizations

$R_{*}$: Rate of Habitable Star Formation per Year

$f_{p}$: Fraction of Stars with Planets

$n_{e}$: Number of Planets Suitable for Life per System

$f_{L}$: Fraction of Habitable Planets where Life Forms

$f_{i}$: Fraction of Planets with Life where Intelligence Develops

$f_{c}$: Fraction of Civilizations with Detectable Technology

$L$: Average Length a Civilization Will Be Detectable

Many of the parameters in the Drake equation are, at best, estimates. Since the initial publishing of the equation our estimates of these values have improved, with some studies finding a value for N as low as 1(Seager, 2018). From this number, it seems incredibly unlikely that any extraterrestrial society is visiting the planet. ## Analysis 

To sufficiently address this guiding question, appropriate analysis must be conducted. With the use of Python and a collection of respective libraries (NumPy, Pandas, Matplotlib, and Scipy), one begins to encroach upon a clear understanding of the relationship 
between economic factors and UFO sightings. To formulate this understanding, three visualizations shall be utilized for the analysis: a correlation heatmap, bar-line chart, and scatterplot. 

Read in both CSV files with Pandas data read_csv() function.

```python
def read_data_files_s4():
    econ_file_s4 = pd.read_csv("Econfile.csv") 
    ufo_file_s4 = pd.read_csv("UFO_sightings_data.csv")

    return econ_file_s4, ufo_file_s4
```

Prior to beginning our analysis, it is crucial that both the ECON and UFO datasets be sufficiently prepared. To do so, a variety of cleaning techniques were utilized before ultimately joining the two datasets. To begin this process, we first eliminated any “NAs” present in the dataset through the utilization of the pandas “.dropna()” function; mitigating the risk for potential NA complications when running certain computations or other operations.

Once this process has been executed, one can direct their focus to the ECON dataset. As we will be joining both datasets via “date”, it is crucial that we adjust the UFO date column to a “DateTime” object to allow sufficient interaction. One can then only retain the year in the date column via “dt.year”. Mortgage rates, unemployment rates, Consumer Price Index (CPI), and disposable income will be the relevant columns for the sake of analysis, other columns (NASDAQ, personal consumption expenditure, and personal savings) will be eliminated from the analysis. The NASDAQ will be eliminated as the S&P 500 is traditionally used as a market indicator proxy; the NASDAQ is characterized by high growth – leading to a bias analysis characterized by obscene high growth, not representative of typical financial markets. Moreover, personal savings and consumption expenditures were eliminated as these variables have a conflicting overlap with the more “boiled down” metric of disposable income. To collect the desired columns, an empty data frame was crafted in which a harmonic mean for each of the above-desired columns (grouped by year via Pandas function “.groupby”) was computed and appended to the ‘econ_dataframe_result’ data frame. The resulting and finalized ECON data frame contained the annual harmonic mean for each economic factor for each year (see below). Note that data points were aggregated to an annual increment for the sake of identifying long-term trends and avoiding unnecessary or misleading “noise” in both the correlation heatmap and bar-line chart caused by monthly granularity. Moreover, annual harmonic means were utilized via the ‘scipy.stats’ library to assign a lesser weighting to outliers and avoid skewing the data, providing a more accurate mean value. The resulting data frame spanned a duration of 42 years (1980-2022).

Looking to the UFO dataset preparation component, a data frame consisting of the year titled “date” and the annual count of UFO sightings across the United States (two columns) was desired. One observes that a pre-existing “Year” column is already present; thus the column titled “Year” was changed to “date” to assist in joining both datasets. To adjust the dataset sufficiently for the sake of analysis, and create a new data frame with the desired information, the dataset was filtered for country code “USA” and grouped by “date” (year) via “.sum()” to provide an annual count of UFO sightings per year. The resulting final data frame “ufo_dataframe_result” was produced for a duration of 104 years (1910 to 2014)(see below).

The above two datasets were finally merged via the pandas “.merge()” function by the “date” column and the duration/ timeframe was transformed automatically per the intersection of 1980 to 2014 (see below) for a final dataset timeframe of 34 years.

```python
head_count = 10
```

```python
econ_file_s4, ufo_file_s4 = read_data_files_s4()

def joint_clean_s4(econ_file_s4, ufo_file_s4,head_count):

    #drop NA's
    econ_file_s4.dropna()
    ufo_file_s4.dropna()

    ###Econ file###
    orignal_econ_date_format_s4 = econ_file_s4['date'] #store orig format for later
    econ_file_s4['date'] = pd.to_datetime(econ_file_s4['date']).dt.year # set date to year


    econ_dataframe_result_s4 = pd.DataFrame() #create empty dataframe
    desired_columns_econ_s4 = ['Mortgage_rate', 'Unemp_rate', 'CPI','disposable_income'] #state desired columns
    for i in range(len(desired_columns_econ_s4)): # take harmonic means for each year
        harmoic_mean_s4 = econ_file_s4.groupby('date')[f'{desired_columns_econ_s4[i]}'].agg(hmean) #annual harmoci mean
        econ_dataframe_result_s4[f'{desired_columns_econ_s4[i]}'] = harmoic_mean_s4


    ###Ufo file###
    ufo_file_s4.rename(columns={'Year': "date"}, inplace=True) #rename index
    ufo_file_s4.set_index('date', inplace=True) # set index 
    ufo_file_s4 = pd.DataFrame(ufo_file_s4[ufo_file_s4['Country_Code']=='USA'].groupby('date').size())
    ufo_file_s4.rename(columns={0 : 'Annual UFO Sighting Count'}, inplace=True) # name count column
    ufo_dataframe_result_s4 = ufo_file_s4


    #merge both
    econ_ufo_merge_df_s4 = pd.merge(ufo_dataframe_result_s4,econ_dataframe_result_s4, on='date')

    #strip last year with half data 
    ###econ_ufo_merge_df_s4 = econ_ufo_merge_df_s4.drop(2014, axis=0)

    return econ_ufo_merge_df_s4, orignal_econ_date_format_s4

econ_ufo_merge_df_s4, orignal_econ_date_format_s4 = joint_clean_s4(econ_file_s4, ufo_file_s4,head_count)
```

The preparation and cleaning phase of the analysis resulted in a sufficient dataset to address the guiding question above. When determining how the economic climate connects to the frequency of UFO sightings across the United States, a correlation 
heatmap was first utilized to determine the relationship between the four chosen economic factors (Mortgage rates, unemployment rates, CPI, and disposable income) and UFO annual sightings. The resulting visualization was as follows: 

Correlation Matrix utilizes dataframe "econ_ufo_merge_df"

```python
def correlation_matrix_map_s4(econ_ufo_merge_df_s4):
    #corr table
    econ_ufo_corr_s4 = econ_ufo_merge_df_s4.corr()
    print(econ_ufo_corr_s4)

    #viz
    plt.matshow(econ_ufo_corr_s4, cmap='coolwarm', fignum=1)
    plt.colorbar()
    plt.xticks(range(len(econ_ufo_corr_s4.columns)),econ_ufo_corr_s4.columns, fontsize=5)
    plt.yticks(range(len(econ_ufo_corr_s4.columns)), econ_ufo_corr_s4.columns, fontsize=5)
    plt.title(" Correlation | UFO Sighting Frequency to ECON Factors")
    plt.show()

correlation_matrix_map_s4(econ_ufo_merge_df_s4)
```

Throughout the component of the analysis in which we direct focus toward correlation, it is crucial to keep in mind that correlation is not always a function of causation. Observed above, we first notice that unemployment is neutrally correlated to the annual frequency of UFO sightings (indicating there is little to no relationship between the two factors) at a correlation value of 0.025. Prior to analysis, a relationship between unemployment and UFO sightings not only seemed promising but certain at face value; however, within the scope of the analysis, we do not see a clear relationship. Looking to mortgage rates and their relationship to UFO sighting frequency on an annual harmonic mean basis, we see a strong negative correlation of -0.7713. this indicates that as mortgage rates and UFO sightings are inversely related in the context of the data. Although one could argue that a decrease in mortgage rates leads to an increase in UFO sightings, it is crucial to note that mortgage rates have declined from their double-digit highs in the 1980s (caused by a lack of mortgage credit suppliers) due to increased liquidity and availability of Mortgage-backed securities (MBS) and mortgage providers. Thus, it is tough to pinpoint if decreased mortgage rates are the cause of increased UFO sightings or mortgages (or vice versa), among much more statistically significant factors are responsible for the increase in UFO sightings. Lastly, looking at the consumer price index (CPI – more informally known as a staple inflation measure) and disposable income, we see both possess a strong correlation with one another and UFO sighting frequency; with each other: 0.9938, with UFO sighting frequency: 0.8902 (CPI) and 0.9075 (Disposable income). The strong correlation between CPI and disposable income is easily explainable as the psychological element of inflation (in which higher prices cause workers to demand higher wages) easily explains an increase in disposable income. As CPI and disposable income create a positive feedback loop with one another and both perpetually increase over time from the perspective of a long-term horizon it is logical that both increase over the long-term time horizon. When directing one's focus to the correlation between the two variables and increase in UFO sightings, we see that both CPI and disposable income have a strong correlation to UFO sightings (also increasing over time). It is especially difficult to pinpoint if an increase in inflation and/or disposable income has played a significant role in the increase in UFO sightings (compared to other factors such as popularity of certain sci-fi films) – none the less, it is evident that there is a significant relationship between mortgage rates, CPI, and disposable income in relation to UFO sightings.

Further delving into the notion of how the economic climate connects to the frequency of UFO sightings across the United States, we can look to the second visualization of the line-bar chart. Matplot, in tandem with Pandas was utilized to create a subplot consisting of four bar-line charts to compare each economic factor to an increase in UFO sightings. Similarly echoing the sentiment from the correlation heatmap, we see that mortgage rates travel inversely to an increase in UFO sightings for the 34-year duration. Similarly, in unifying fashion to the correlation heatmap, we see that CPI and disposable income travel in the same direction as the frequency of UFO sightings. Moreover, once again, unemployment does not follow any form of trend in relation to UFO sighting frequency; granted when unemployment is at its highest point for 1980-2014, UFO sightings also reach their all-time high within the same timeframe. To view these observations, please see below.

Bar-line utilizes econ_ufo_merge_df

```python
def bar_map_s4(econ_ufo_merge_df_s4):
    #define quadrants
    fig, ax = plt.subplots(2,2,figsize=(10, 8))
    #ufo and mort
    ax[0,0].bar(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['Annual UFO Sighting Count'], color='red')
    ax21 = ax[0,0].twinx()
    ax21.plot(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['Mortgage_rate'], label='Mortgage Rates(%)')
    plt.grid()
    plt.legend()

    ax[0,1].bar(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['Annual UFO Sighting Count'], label='Unemployment Rate(%)',color='red')
    ax22 = ax[0,1].twinx()
    ax22.plot(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['Unemp_rate'], label='Unemployment Rate(%)')
    plt.grid()
    plt.legend()

    ax[1,0].bar(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['Annual UFO Sighting Count'], label='CPI',color='red')
    ax23 = ax[1,0].twinx()
    ax23.plot(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['CPI'], label='CPI')
    plt.grid()
    plt.legend()

    ax[1,1].bar(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['Annual UFO Sighting Count'], label='Disposable Income',color='red')
    ax24 = ax[1,1].twinx()
    ax24.plot(econ_ufo_merge_df_s4.index, econ_ufo_merge_df_s4['disposable_income'], label='Disposable Income')
    plt.grid()
    plt.legend()

    fig.suptitle("Annual Seasonality of UFO Sightings to Economic Factors",fontsize=16, y=0.98)
    plt.tight_layout()

bar_map_s4(econ_ufo_merge_df_s4)
```

Glancing to the last component of the analysis to sufficiently answer how the economic climate connects to the frequency of UFO sightings across the United States, one looks to the scatter plot component of the analysis. To ensure a more appropriate sample size for a scatterplot metric, the analysis will be shifted from annual to monthly to boost granularity from 34 to 402 data points. Note that harmonic means still apply when applicable and a monthly sum has been applied for the count of UFO sightings per month. This was achieved by once again filtering for the country code ‘USA’ and grouping by ‘Date’ which was presented in a year-month format (yyyy-mm).

Scatter requires separate cleaning mechanism to boost granularity from annual to monthly data

references used: https://www.geeksforgeeks.org/pandas-series-dt-to_period/ rational: to_period('M') to retain only month

```python
def scatter_map_s4(econ_file_s4, ufo_file_s4,orignal_econ_date_format_s4,head_count):
    #ECON file manipulation
    print(econ_file_s4)
    econ_file_s4['date'] = orignal_econ_date_format_s4 #revert econ date column to orignal monthly data to boost granularity
    econ_file_s4['date'] = pd.to_datetime(econ_file_s4['date']).dt.to_period('M') #get to yyyy-mm 
    econ_file_s4 = econ_file_s4[['date','Mortgage_rate', 'Unemp_rate', 'CPI','disposable_income']] #isolate desired columns #no harmoc mean required as monmthly is base increment

    #UFO file manipulation 
    ufo_file_s4 = ufo_file_s4.reset_index(drop=True)
    ufo_file_s4['date'] = pd.to_datetime(ufo_file_s4['Date_time']).dt.to_period('M') # create new column
    ufo_file_s4 = pd.DataFrame(ufo_file_s4[ufo_file_s4['Country_Code']=='USA'].groupby('date').size()) # get monthly count
    ufo_file_s4.rename(columns={0: 'Monthly UFO Sighting Count'}, inplace=True)

    #merge ECON & UFO
    econ_ufo_merge_granular_df_s4 = pd.merge(econ_file_s4,ufo_file_s4, on='date')

    #strip last year
    ####econ_ufo_merge_granular_df_s4 = econ_ufo_merge_granular_df_s4.iloc[:-5]

    return econ_ufo_merge_granular_df_s4

econ_ufo_merge_granular_df_s4 = scatter_map_s4(econ_file_s4, ufo_file_s4,orignal_econ_date_format_s4,head_count)
```

When one observes the four scatter plots below, it is crucial to observe two key factors for each; relationship directionality and strength of that relationship (also known as ‘r’ or the correlation coefficient – indicated by how closely the points cluster together along the trend line). In relation to monthly UFO sighting count, we see that mortgage rates possess a relatively strong negative relationship (mortgage rates and sighting frequency often travel inversely to one another) displayed by the downward trend with data points remaining close to the line of best fit. In Juxtaposition to mortgage rates, we see that both CPI and disposable income have a very positive relationship with UFO sighting frequency; indicated by an upward trend with data points that closely cluster the trend line. Lastly, one observes a weak neutral relationship between unemployment and UFO sighting frequency indicated by loosely scattered points that fail to gravitate toward the trend line. One can further support this claim via the weak correlation coefficient between the two variables and the lack of symmetry present in the bar-line chart.

A noteworthy occurrence takes place in the scatter plots fo, CI, and disposable income. As we travel along the trend line, regardless of directionality, we see that the direction of the relationship does not change but the strength does, as we see an increase in UFO sightings, along with an increase in the underlying variable, the see the points disperse further from the trend line; indicating potential heteroscedasticity or an external influence on the variables in question within each scatterplot.

Scatter utilizes econ_ufo_merge_granular_df_s4

```python
def scatter_map_vis_s4(econ_ufo_merge_granular_df_s4):

    #vis
    fig, ax = plt.subplots(2,2,figsize=(10, 8))

    ax[0, 0].scatter(econ_ufo_merge_granular_df_s4['Mortgage_rate'], econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'], color='red', label='Mortgage Rates(%)', s=0.9)
    ax[0, 0].set_ylabel('UFO Sightings')
    ax[0, 0].set_xlabel('Mortgage Rates(%)')
    ax[0, 0].set_title('Mortgage Rates relative to UFO Sightings')
    ax[0, 0].grid()
    m1,b1 = np.polyfit(econ_ufo_merge_granular_df_s4['Mortgage_rate'],econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'],1)
    ax[0,0].plot(econ_ufo_merge_granular_df_s4['Mortgage_rate'], m1*econ_ufo_merge_granular_df_s4['Mortgage_rate'] + b1, color='black')

    ax[1, 0].scatter(econ_ufo_merge_granular_df_s4['Unemp_rate'], econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'], color='blue',s=0.9)
    ax[1, 0].set_ylabel('UFO Sightings')
    ax[1, 0].set_xlabel('Unemployment Rate(%)')
    ax[1, 0].set_title('Unemployment Rate relative to UFO Sightings')
    ax[1, 0].grid()
    m2,b2 = np.polyfit(econ_ufo_merge_granular_df_s4['Unemp_rate'],econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'],1)
    ax[1,0].plot(econ_ufo_merge_granular_df_s4['Unemp_rate'], m2*econ_ufo_merge_granular_df_s4['Unemp_rate'] + b2, color='black')

    ax[0, 1].scatter(econ_ufo_merge_granular_df_s4['CPI'], econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'], color='green',s=0.9)
    ax[0, 1].set_ylabel('UFO Sightings')
    ax[0, 1].set_xlabel('CPI Index')
    ax[0, 1].set_title('CPI Index relative to UFO Sightings')
    ax[0, 1].grid()
    m3,b3 = np.polyfit(econ_ufo_merge_granular_df_s4['CPI'],econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'],1)
    ax[0,1].plot(econ_ufo_merge_granular_df_s4['CPI'], m3*econ_ufo_merge_granular_df_s4['CPI'] + b3, color='black')

    ax[1, 1].scatter(econ_ufo_merge_granular_df_s4['disposable_income'], econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'], color='green',s=0.9)
    ax[1, 1].set_ylabel('UFO Sightings')
    ax[1, 1].set_xlabel('Disposable Income')
    ax[1, 1].set_title('Disposable Income relative to UFO Sightings')
    ax[1, 1].grid()
    m4,b4 = np.polyfit(econ_ufo_merge_granular_df_s4['disposable_income'],econ_ufo_merge_granular_df_s4['Monthly UFO Sighting Count'],1)
    ax[1,1].plot(econ_ufo_merge_granular_df_s4['disposable_income'], m4*econ_ufo_merge_granular_df_s4['disposable_income'] + b4, color='black')

    fig.suptitle("UFO Sightings to Economic Factors", fontsize=16, y=0.98)
    plt.tight_layout()

scatter_map_vis_s4(econ_ufo_merge_granular_df_s4)
```
